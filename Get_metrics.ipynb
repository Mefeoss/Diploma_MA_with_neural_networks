## Levenshtein distance

Определим функцию для подсчета процента схожести строк:

```python
def similarity_percentage_lev(s1, s2):
    if len(s1) == 0:
        return 0.0
    
    distance = [[0] * (len(s2) + 1) for _ in range(len(s1) + 1)]

    for i in range(len(s1) + 1):
        distance[i][0] = i

    for j in range(len(s2) + 1):
        distance[0][j] = j

    for i in range(1, len(s1) + 1):
        for j in range(1, len(s2) + 1):
            cost = 0 if s1[i - 1] == s2[j - 1] else 1
            distance[i][j] = min(distance[i - 1][j] + 1,        
                                 distance[i][j - 1] + 1,        
                                 distance[i - 1][j - 1] + cost) 

    max_length = max(len(s1), len(s2))
    similarity = 1.0 - (distance[len(s1)][len(s2)] / max_length)
    return similarity

def average_similarity_percentage_lev(pairs):
    total_similarity = 0.0
    pair_count = len(pairs)

    for pair in pairs:
        s1, s2 = pair
        similarity = similarity_percentage_lev(s1, s2)
        total_similarity += similarity

    average_similarity = total_similarity / pair_count
    return average_similarity
```

```python
def average_similarity_lev_words(pairs):
    total_similarity = 0.0
    n = 0

    for pair in pairs:
        s1, s2 = pair
        splited_s1 = s1.split(',')
        splited_s2 = s2.split(',')
        if len(splited_s1) == len(splited_s2):
          for i in range(0,len(splited_s1)):
            similarity = similarity_percentage_lev(splited_s1[i], splited_s2[i])
            total_similarity += similarity
            n+=1
        else:
          similarity = similarity_percentage_lev(s1, s2)
          total_similarity += similarity
          n+=1

    average_similarity = total_similarity / n
    return average_similarity
```

## Sequence Matcher

```python
from difflib import SequenceMatcher

def similarity_percentage_seq(s1, s2):
    matcher = SequenceMatcher(None, s1, s2)
    similarity = matcher.ratio()
    return similarity

def average_similarity_seq(pairs):
    total_similarity = 0.0
    pair_count = len(pairs)

    for pair in pairs:
        s1, s2 = pair
        similarity = similarity_percentage_seq(s1, s2)
        total_similarity += similarity

    average_similarity = total_similarity / pair_count
    return average_similarity
```

```python
def average_similarity_seq_words(pairs):
    total_similarity = 0.0
    n = 0

    for pair in pairs:
        s1, s2 = pair
        splited_s1 = s1.split(',')
        splited_s2 = s2.split(',')
        if len(splited_s1) == len(splited_s2):
          for i in range(0,len(splited_s1)):
            similarity = similarity_percentage_seq(splited_s1[i], splited_s2[i])
            total_similarity += similarity
            n+=1
        else:
          similarity = similarity_percentage_seq(s1, s2)
          total_similarity += similarity
          n+=1

    average_similarity = total_similarity / n
    return average_similarity
```

```python
average_similarity_half2 = average_similarity_seq(half_pairs)
print(average_similarity_half2)
```

```python
average_similarity_full2 = average_similarity_seq(full_pairs)
print(average_similarity_full2)
```

```python
aver_sim_half2 = average_similarity_seq_words(half_pairs)
print(aver_sim_half2)
```

```python
aver_sim_full2 = average_similarity_seq_words(full_pairs)
print(aver_sim_full2)
```

## Cosine similarity

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def similarity_percentage_cos(s1, s2):
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform([s1, s2])
    similarity = cosine_similarity(vectors[0], vectors[1])[0][0]
    return similarity

def average_similarity_cos(pairs):
    total_similarity = 0.0
    pair_count = len(pairs)

    for pair in pairs:
        s1, s2 = pair
        similarity = similarity_percentage_cos(s1, s2)
        total_similarity += similarity

    average_similarity = total_similarity / pair_count
    return average_similarity
```

```python
average_similarity_half3 = average_similarity_cos(half_pairs)
print(average_similarity_half3)
```

```python
average_similarity_full3 = average_similarity_cos(full_pairs)
print(average_similarity_full3)
```

## Jaccard similarity

```python
def similarity_percentage_jac(s1, s2):
    set1 = set(s1)
    set2 = set(s2)
    intersection = len(set1 & set2)
    union = len(set1 | set2)
    similarity = intersection / union if union != 0 else 0.0
    return similarity

def average_similarity_jac(pairs):
    total_similarity = 0.0
    pair_count = len(pairs)

    for pair in pairs:
        s1, s2 = pair
        similarity = similarity_percentage_jac(s1, s2)
        total_similarity += similarity

    average_similarity = total_similarity / pair_count
    return average_similarity
```

```python
def average_similarity_jac_words(pairs):
    total_similarity = 0.0
    n = 0

    for pair in pairs:
        s1, s2 = pair
        splited_s1 = s1.split(',')
        splited_s2 = s2.split(',')
        if len(splited_s1) == len(splited_s2):
          for i in range(0,len(splited_s1)):
            similarity = similarity_percentage_jac(splited_s1[i], splited_s2[i])
            total_similarity += similarity
            n+=1
        else:
          similarity = similarity_percentage_jac(s1, s2)
          total_similarity += similarity
          n+=1

    average_similarity = total_similarity / n
    return average_similarity
```

## Jaro-winkler
